```python
import sys
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

## Definir les sommets et les aretes

```python
G= nx.DiGraph()

edges = [('19h','20h',30), ('19h','23h',50), ('19h','22h',40), ('19h','21h',30),
         ('20h','00h',50), ('20h','22h',30), ('20h','23h',40), ('21h','23h',30),
         ('21h','01h',50), ('21h','00h',40), ('22h','00h',30), ('22h','01h',40),
         ('23h','01h',30), ('23h','02h',40), ('00h','02h',30), ('01h','02h',30),]
```

```python
G.add_weighted_edges_from(edges)
pos=nx.circular_layout(G)
plt.figure(figsize=(10,5))
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels,font_color='blue')
nx.draw(G,pos,with_labels=True,
        node_size=700,node_color='orange',
        width=2,linewidths=2,alpha=1)
plt.show()
```

## Le chemin le plus court entre 19h et 02h

```python
nx.dijkstra_path(G, '19h', '02h')
```
